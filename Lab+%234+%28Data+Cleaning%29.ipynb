### Candy Data

This is the first data set: 

https://boingboing.net/2016/10/31/the-candy-hierarchy-for-2016.html

It is called "BOING BOING CANDY HIERARCHY 2016 SURVEY Responses" and is data for the Candy hierarchy survey from 2016.

With this data set, my goal was to try and clean the .head() block of data as much as I could, with the time and knowlegde given. 

```python
#This is one line to open and read my csv file into a data frame data structure
#Additionally, I needed a special encoding add-in because my computer was blowing up
#Ms.Sconyers helped me find the specfic encoding that I needed
#I set this data set as 'df' because it was the first of three data sets
df = pd.read_csv('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.csv', encoding= 'ISO-8859-1')
```

```python
#Here, I gave the first head() columns new column names
#I renamed them using the data cleaning cheat sheet code
# I also used code Ms.Sconyers shared in class by using index calling rather than renaming the original name

df.rename(columns={
    'Timestamp': 'time',
    'Are you going actually going trick or treating yourself?': 'trick_or_treat',
    'Your gender:': 'gender',
    'How old are you?':'age',
    'Which country do you live in?':'country',
    'Which state, province, county do you live in?' : 'state',
    '[100 Grand Bar' : '100_grand_bar',
    list(df)[6] : '100_grand_bar',
    '[Anonymous brown globs that come in black and orange wrappers]' : 'globs',
    list(df)[7] : 'globs',
    '[Any full-sized candy bar]' : 'full_size',
    list(df)[8] : 'full_size',
    '[Black Jacks]' : 'black_jack',
    list(df)[9] : 'black_jack',
    list(df)[112] : 'font',
    list(df)[113] : 'rowling',
    list(df)[114] : 'jj_abrams',
    list(df)[115] : 'beyonce',
    list(df)[116] : 'bieber',
    list(df)[117] : 'k_bacon',
    list(df)[118] : 'f_bacon',
    list(df)[119] : 'day',
    list(df)[120] : 'apples',
    list(df)[121] : 'news'},
    inplace=True)
df.head()
```

```python
#Next, I dropped the last column that said to the ignore data in the column header
#I did this using the .drop() function in pandas and called the column header name in order to do this
#I also had to call the axis and the boolean function "True" to drop the entire header
df.drop(' [York Peppermint Patties] Ignore', axis = 1, inplace = True)
```

```python
#Here, I made all the items in a few select columns upper case with the .upper() fxn
#I did this to better read the data in the column and make it all the same format
#I also put some columns with an uppercase for first letter with the .title() fxn
#I did this to also better read the data in the column and each item the same format
df['country'] = df['country'].str.upper()
df['state'] = df['state'].str.upper()
df['font'] = df['font'].str.title()
df.head()
```

```python
#Next, I used the .replace() funtion to change some of the incorrect data input
#I did this using the .replace() fxn by calling the original name and overriding it with a new name
#I changed the meh-joy-depair options to 1-2-3 format to simplify the vast amount of meh-joy-despair data
#I also changed the state and province names to un-abbreviated as to format the code in that column the same
#Making each item of these columns the same format (in the previous block of code) helped me simplify my code 
df = df.replace(['MEH', 'JOY', 'DESPAIR','US','IL'], [1,2,3,'USA','ILLINOIS'])
df.head()
```
